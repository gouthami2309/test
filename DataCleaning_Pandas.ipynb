```python
import pandas as pd
```

```python
df= pd.read_excel(r"/content/Customer Call List.xlsx")
df
```

```python
df= df.drop_duplicates()
df
```

```python
df= df.drop(columns= "Not_Useful_Column")
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip("123./")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[:3]+'-'+ x[3:6]+'-'+ x[6:])
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--',' ')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--',' ')
df
```

```python
df[["Street_Address","State","Zip_Code"]]=df["Address"].str.split(',', n=2, expand=True)
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes', 'Y', regex=True)
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No', 'N', regex=True)
df

```

```python
df=df.replace('N/a','')

df= df.fillna('')
df
```

```python
for x in df.index:
  if df.loc[x,"Do_Not_Contact"]=='Y':
    df.drop(x,inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"]=='--':
    df.drop(x,inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"]==' ':
    df.drop(x,inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```
